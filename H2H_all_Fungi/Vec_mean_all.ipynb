```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from glob import glob
```

```python
# Obtener el directorio actual
directorio_actual = os.getcwd()

print("El directorio actual es:", directorio_actual)
```

```python
# Obtener la lista de archivos gff
files_vec = glob("*vecinos.csv")
print(files_vec)
# Lista para almacenar el número de filas de cada archivo
num_rows = []

# Iterar sobre cada archivo gff
for file in files_vec:
    try:
        # Leer el archivo gff
        df_gff = pd.read_csv(file, sep=",")
        
        # Contar el número de filas y añadirlo a la lista
        num_rows.append(len(df_gff))
        
    except pd.errors.ParserError:
        # Manejar el error (por ejemplo, imprimir un mensaje, omitir el archivo)
        print(f"Error al analizar el archivo: {file}")

# Crear un DataFrame con los nombres de los archivos y el número de filas
df_num_rows = pd.DataFrame({"Archivo": [os.path.splitext(file)[0] for file in files_vec], "Num_filas": num_rows})

# Ordenar el DataFrame por número de filas
df_num_rows = df_num_rows.sort_values(by="Num_filas", ascending=False)
df_num_rows
```

```python
# Crear el gráfico de barras con Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="Archivo", y="Num_filas", data=df_num_rows, palette="viridis")

# Añadir título y etiquetas a los ejes
plt.title('Número de filas por archivo gff')
plt.xlabel('Archivo')
plt.ylabel('Número de filas')

# Rotar las etiquetas del eje x para una mejor visualización
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.show()
```

```python
vecinos_csv = glob("*_vecinos.csv")
for file in vecinos_csv:
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv(file, sep=",", skiprows=1, header=None)
    df =df.drop(columns=[1])
# Guardar el DataFrame como un archivo de texto separado por tabulaciones (TXT)
# Obtener el nombre del archivo sin la extensión y separar por guiones
    nombre_archivo = os.path.splitext(file)[0]
    nombre_split = nombre_archivo.split('_')
        # Obtener el primer elemento de nombre_split
    nombre_final = nombre_split[0]

        # Guardar el DataFrame modificado en un nuevo archivo .csv con el nombre deseado
    df.to_csv(f"{nombre_final}_gff.txt", sep="\t",index=False)
```
